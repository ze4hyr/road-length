```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import requests
```

```python
data_folder = 'data'
output_folder = 'output'

if not os.path.exists(data_folder):
    os.mkdir(data_folder)
if not os.path.exists(output_folder):
    os.mkdir(output_folder)
```

```python
def download(url):
    filename = os.path.join(data_folder, os.path.basename(url))
    if not os.path.exists(filename):
      with requests.get(url, stream=True, allow_redirects=True) as r:
          with open(filename, 'wb') as f:
              for chunk in r.iter_content(chunk_size=8192):
                  f.write(chunk)
      print('Downloaded', filename)
```

```python
files = [
  '2020-01-metropolitan-street.csv',
  '2020-02-metropolitan-street.csv',
  '2020-03-metropolitan-street.csv',
  '2020-04-metropolitan-street.csv',
  '2020-05-metropolitan-street.csv',
  '2020-06-metropolitan-street.csv',
  '2020-07-metropolitan-street.csv',
  '2020-08-metropolitan-street.csv',
  '2020-09-metropolitan-street.csv',
  '2020-10-metropolitan-street.csv',
  '2020-11-metropolitan-street.csv',
  '2020-12-metropolitan-street.csv'
]


data_url = 'https://github.com/spatialthoughts/python-dataviz-web/releases/' \
  'download/police.uk/'

for f in files:
  url = os.path.join(data_url + f)
  download(url)
```

```python
dataframe_list = []

for f in files:
    filepath = os.path.join(data_folder, f)
    df = pd.read_csv(filepath)
    dataframe_list.append(df)

merged_df = pd.concat(dataframe_list)
merged_df
```

```python
type_counts=merged_df.groupby('Crime type').size()
print(type_counts)
```

```python
fig, ax=plt.subplots(1,1)
fig.set_size_inches(15, 10)
type_counts.plot(kind='pie', ax=ax,
                 wedgeprops={'linewidth': 3, 'edgecolor': 'white'},
                 textprops={'fontstyle': 'italic', 'fontsize': 10, 'fontfamily': 'Calibri'})
ax.set_title('Crime Types', fontsize=18, fontweight='bold')
ax.set_ylabel('')
plt.tight_layout
plt.show()
```

```python
month_count=merged_df.groupby('Month').size()
month_count
```

```python
fig, ax=plt.subplots(1,1)
fig.set_size_inches(15, 10)
month_count.plot(kind='bar', ax=ax,
                  color='orange',
                  edgecolor='black',
                  linewidth=2)
month_count.plot(kind='line', ax=ax,
                  color='red',
                  linewidth=2,
                  marker='o',
                  markersize=8)
ax.set_title('Crime Counts by Month', fontsize=18, fontweight='bold')
ax.set_xlabel('Month', fontsize=14, fontweight='bold')
ax.set_ylabel('Crime Count', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
bycycle_thefts=merged_df[merged_df['Crime type']=='Bicycle theft']
bycycle_thefts.head()
```

```python
bycycle_theft_month=bycycle_thefts.groupby('Month').size()
bycycle_theft_month
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")
sns.set_palette(sns.color_palette("dark", n_colors=10))

fig, ax=plt.subplots(1,1)
fig.set_size_inches(15, 10)
bycycle_theft_month.plot(kind='bar', ax=ax,)
ax.set_title('Bicycle Thefts by Month', fontsize=18, fontweight='bold')
ax.set_xlabel('Month', fontsize=14, fontweight='bold')
ax.set_ylabel('Bicycle Theft Count', fontsize=14, fontweight='bold')
for i, value in enumerate(bycycle_theft_month.values):
    ax.annotate(str(value), xy=(i, value), xytext=(0, 5), textcoords='offset points', 
                ha='center', fontsize=12, fontweight='bold', 
                bbox=dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="white"))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
